```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_avaliacoes = pd.read_csv('avaliacoes_servicos.csv', sep=';', encoding='ISO-8859-1')
df_info = pd.read_csv('info_usuarios.csv', sep=";" , encoding='ISO-8859-1')
df_metodo_pagamento = pd.read_csv('metodo_pagamento.csv', sep=';' , encoding='ISO-8859-1')
df_pagamentos = pd.read_csv('pagamentos.csv', sep=';' , encoding='ISO-8859-1')
df_pedidos = pd.read_csv('pedidos.csv', sep=';' , encoding='ISO-8859-1')
df_pedidos_servicos = pd.read_csv('pedidos_servico.csv', sep=';' , encoding='ISO-8859-1')
df_perfil_freelancer = pd.read_csv('perfil_freelancer.csv', sep=';' , encoding='ISO-8859-1')
df_servicos = pd.read_csv('servicos.csv', sep=';' , encoding='ISO-8859-1')
df_pedidos = pd.read_csv('pedidos.csv', sep=';' , encoding='ISO-8859-1')

```

```python
print(df_servicos.columns)
```

```python
# Agrupa por ID do serviço e calcula a média da avaliação
media_avaliacoes = df_avaliacoes.groupby('id_servico')['nota'].mean().reset_index()

# Junta com a tabela de serviços para obter o título do serviço
df_merge = media_avaliacoes.merge(df_servicos, on='id_servico')

# Ordena do maior para o menor
df_merge = df_merge.sort_values(by='nota', ascending=False)
```

# Média de Avaliações dos Serviços


```python

plt.figure(figsize=(10, 6))
plt.bar(df_merge['titulo'], df_merge['nota'], color='skyblue')
plt.xticks(rotation=45, ha='right')
plt.xlabel('Serviço')
plt.ylabel('Nota média')
plt.title('Média de Avaliações por Serviço')
plt.tight_layout()
plt.show()
```
